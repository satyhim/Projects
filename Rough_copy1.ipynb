```python
NOW = dt.datetime(2011,12,10)
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
RFM_data=df.groupby('CustomerID').agg({'InvoiceDate':lambda x:(NOW-x.max()).days,'InvoiceNo':lambda x:len(x),'Total_price':lambda x: x.sum()})
RFM_data.head(10)
```

```python
RFM_data.rename(columns={'InvoiceDate':'Recency','InvoiceNo':'Frequency','Total_price':'monetary_value'},inplace=True)
RFM_data.head()
```

```python
RFM_data.describe()
```

```python
RFM_data["Recency_Score"]= pd.qcut(RFM_data["Recency"],5,labels=[5,4,3,2,1])
RFM_data.head()
```

```python
RFM_data.describe()
```

```python
RFM_data["Frequency_Score"]= pd.qcut(RFM_data["Frequency"].rank(method="first"),5, labels=[1,2,3,4,5])
RFM_data.head()
```

```python
RFM_data["Monetary_Score"]=pd.qcut(RFM_data["monetary_value"],5, labels=[1,2,3,4,5])
RFM_data.head()
```

```python
RFM_data["RFM"]=RFM_data["Recency_Score"].astype(str)+RFM_data["Frequency_Score"].astype(str)
RFM_data.head(10)
```

```python
RFM_data['Segment'] = RFM_data['RFM']
RFM_data.head()
```

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}
```

```python
RFM_data['Segment'] = RFM_data['Segment'].replace(seg_map, regex=True)
RFM_data.reset_index(inplace=True)
RFM_data.head()
```

```python
RFM_data.groupby('Segment').agg({"CustomerID":"count"}).sort_values("CustomerID",ascending=False)
```

```python
colors  = ("darkorange", "darkseagreen", "orange", "cyan", "cadetblue", "hotpink", "lightsteelblue", "coral",  "mediumaquamarine","palegoldenrod")
explodes = [0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25]

RFM_data["Segment"].value_counts(sort=False).plot.pie(colors=colors,
                                                 textprops={'fontsize': 12}, 
                                                 autopct = '%4.1f',
                                                 startangle= 90, 
                                                 radius =2, 
                                                 rotatelabels=True,
                                                 shadow = True, 
                                                 explode = explodes)
plt.ylabel("");
```
